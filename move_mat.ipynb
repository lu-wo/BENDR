```python
import os, sys
```

```python
ROOT = "/Users/lukas/Desktop/projects/deepeye_all/BENDR_emg/data/01_raw"
```

```python
# iterate recursively through all files in ROOT and copy *.mat files to ROOT/mat
for root, dirs, files in os.walk(ROOT):
    for file in files:
        if file.endswith(".mat"):
            print(os.path.join(root, file))
            os.rename(os.path.join(root, file), os.path.join(ROOT, "mat", file))
```
